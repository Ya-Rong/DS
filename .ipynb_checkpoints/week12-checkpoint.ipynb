## Boston房價進行迴歸實作

```python
from sklearn.datasets import load_boston
import pandas as pd
import statsmodels.api as sm

# 載入波士頓房價數據集
boston = load_boston()

# 將資料轉換成 DataFrame 格式
df = pd.DataFrame(boston.data, columns=boston.feature_names)

# 將標籤（房價）加入 DataFrame
df['MEDV'] = boston.target

# 簡單線性回歸（以 RM 為自變數，MEDV 為應變數）
X = df['RM']
y = df['MEDV']
X = sm.add_constant(X)  # 加入截距項
model = sm.OLS(y, X).fit()  # 建立簡單線性回歸模型
print(model.summary())  # 輸出迴歸結果
```

```python
# 多元線性回歸（以所有變數為自變數，MEDV 為應變數）

X = df.drop('MEDV', axis=1)
y = df['MEDV']
X = sm.add_constant(X)  # 加入截距項
model = sm.OLS(y, X).fit()  # 建立多元線性回歸模型
print(model.summary())  # 輸出迴歸結果
```

## 課堂練習



```python
# 計算相關係數與繪製散佈圖 scatter plot

from sklearn.datasets import load_boston
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt

# 載入波士頓房價數據集
boston = load_boston()

# 將資料轉換成 DataFrame 格式
df = pd.DataFrame(boston.data, columns=boston.feature_names)

# 將標籤（房價）加入 DataFrame
df['MEDV'] = boston.target

# 簡單線性回歸（以 RM 為自變數，MEDV 為應變數）
X = df['RM']
y = df['MEDV']

x = np.array(X)
y = np.array(y)

plt.scatter(x, y)
plt.show()
```

```python
# 簡單線性迴歸（以 CRIM 為自變數，MEDV 為應變數）
sns.regplot(x='CRIM', y='MEDV', data=df)
plt.title('Simple Linear Regression')
plt.xlabel('CRIM')
plt.ylabel('MEDV')
plt.show()
```

```python
# 迴歸方程式

# 擷取自變項和依變項
X = df['CRIM']
y = df['MEDV']

X = sm.add_constant(X)  # 加入截距項
model = sm.OLS(y, X).fit()  # 建立簡單線性回歸模型
print(model.summary())  # 輸出迴歸結果

# 計算共變異數和方差
cov_xy = np.cov(x, y, ddof=0)[0, 1]  # 兩個向量的共變異數
var_x = np.var(x, ddof=0)  # x 的方差

# 計算斜率
slope = cov_xy / var_x

# 輸出線性方程式
print(f"線性方程式為: MEDV = X + slope * CRIM ")

# 沒有算出來
```

```python
# 匯入資料
import pandas as pd

df = pd.read_csv("SeoulBikeData.csv", encoding='gbk', index_col=['Date'])
df
```

```python
# 資料清理 (Data cleaning)

data = df.copy()
data = data[data['Functioning Day'] == 'Yes']
data.pop('Functioning Day')
```

```python
# 將文字資料 編碼 為數字

from sklearn.preprocessing import LabelEncoder # 資料預處理
le = LabelEncoder()

data['Seasons'] = le.fit_transform(data['Seasons'])
data['Holiday'] = le.fit_transform(data['Holiday'])

data[['Seasons', 'Holiday']]
# Seasons：0(Spring),1(Summer),2(Autumn),3(Winter)
# Holiday：0(No),1(Yes)
```

```python
# 多元線性回歸（以所有變數為自變數，MEDV 為應變數）
X = df.drop('MEDV', axis=1)
y = df['MEDV']
X = sm.add_constant(X)  # 加入截距項
model = sm.OLS(y, X).fit()  # 建立多元線性回歸模型
print(model.summary())  # 輸出迴歸結果
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# 創建資料信
df = pd.DataFrame({'Area':[10,8,8,5,7,8,7,9,6,9],
                  'Distance':[80,2,200,200,300,230,40,0,330,180],
                  'Sales':[469,366,371,208,246,297,363,436,198,364]})

# 擷取自變項和依變項
X = df[['Hour', 'Temperature', 'Humidity(%)', 'Wind speed(m/s)', 'Visibility(10m)','Dew point temperature', 'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)']]
y = df['Rented Bike Count']

# 建立線性回歸模型
model = LinearRegression()
model.fit(X, y)

# 計算迴歸係數和截距
coef = model.coef_
intercept = model.intercept_

# 輸出線性方程式
print(f"線性方程式為: Sales = {coef[0]:.2f} * Area + {coef[1]:.2f} * Distance + {intercept:.2f}")
```
