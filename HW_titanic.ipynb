```python
import seaborn as sns
import pandas as pd
```

### 

```python
# pclass 1 = 1st, 2 = 2nd, 3 = 3rd alive(yes, no)
df1 = pd.crosstab(titanic["pclass"], titanic["alive"])
df1
```

```python
df1_1 = pd.crosstab(titanic["pclass"], titanic["alive"], normalize="index")
df1_1
```

```python
df1_1.drop(columns=['no'], axis=1, inplace=True)
```

```python
df1_1.plot.bar(stacked=True)
```

### 

```python
# sex alive(yes, no)
df2 = pd.crosstab(titanic["sex"], titanic["alive"])
df2
```

```python
df2_1 = pd.crosstab(titanic["sex"], titanic["alive"], normalize=True)
df2_1
```

```python
df2_1.drop(columns=['no'], axis=1, inplace=True)
```

```python
df2_1.plot.bar(stacked=True)
```

### 

```python
# fare alive(yes, no)
df3 = pd.crosstab(titanic["fare"], titanic["alive"])
df3
```

```python
df3_1 = pd.crosstab(titanic["fare"], titanic["alive"], normalize=True)
df3_1
```

```python
df3_1.drop(columns=['no'], axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(60,6)) # 10x6
ax = df3_1.plot.bar(stacked=True, ax=fig.add_subplot(111))
plt.xticks(rotation=90) # x
plt.subplots_adjust(bottom=0.2) # 
plt.show()
```

### 

```python
# sex fare pclass v.s  alive
df4 = pd.crosstab(index=[titanic["sex"], titanic["fare"], titanic["pclass"]], columns=titanic["alive"])
df4
```

```python
df4_1 = pd.crosstab(index=[titanic["sex"], titanic["fare"], titanic["pclass"]], columns=titanic["alive"], normalize=True)
df4_1
```

```python
df4_1.drop(columns=['no'], axis=1, inplace=True) #  no 
```

```python
# 
import pandas as pd
import numpy as np

# 
df4_1 = df4_1.loc[(df4_1 != 0).any(axis=1)]
# 
print(df4_1)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(60,6)) # 10x6
ax = df4_1.plot.bar(stacked=True, ax=fig.add_subplot(111))
plt.xticks(rotation=90) # x
plt.subplots_adjust(bottom=0.2) # 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#  yes  75% 
q75, qmax = np.percentile(df4_1['yes'], [75, 100])
df4_1 = df4_1[df4_1['yes'].between(q75, qmax)]

fig = plt.figure(figsize=(60,6)) # 10x6
ax = df4_1.plot.bar(stacked=True, ax=fig.add_subplot(111))
plt.xticks(rotation=90) # x
plt.subplots_adjust(bottom=0.2) # 
plt.show()
```
