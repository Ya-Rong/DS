```python
data1 = [800,900,1000,1100,1200]
data2 = [500,750,1000,1250,1500]
print(st.mean(data1))
print(st.mean(data2))
print(st.median(data1))
print(st.median(data2))
```

```python
data1 = [800,900,1000,1100,1200]
data2 = [500,750,1000,1250,1500]
print(st.variance(data1))
print(st.variance(data2))
print(st.stdev(data1))
print(st.stdev(data2))
```
