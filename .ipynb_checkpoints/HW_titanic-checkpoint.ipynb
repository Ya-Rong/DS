```python
import seaborn as sns
import pandas as pd
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic
```

###### survived 是否倖存 0 = No, 1 = Yes
###### pclass 船票等級 1 = 1st, 2 = 2nd, 3 = 3rd
###### sibsp 在船上同為兄弟姐妹或配偶的數目
###### parch 在船上同為家族的父母及小孩的數目
###### ticket 船票編號
###### fare 船票價格
###### embarked 登船的口岸 C = Cherbourg, Q = Queenstown, S = Southampton

```python
df = pd.crosstab(titanic["class"], titanic["sex"])
df
```

```python
# 堆疊長條圖
df.plot.bar(stacked=True) # stacked=True 才會堆疊
```

```python
df.plot.bar(stacked=False) # 橫向堆疊
```

```python
# 利用pandas資料框架會置百分比堆疊長條圖範例
df2 = pd.crosstab(titanic["class"], titanic["sex"], normalize="index")
df2
```

```python
df2.plot.bar(stacked=True)
```

```python
# 堆疊橫條圖
df.plot.barh(stacked=True) # bar 後面加個 h 會改成橫條圖
```

```python
# 堆疊百分比橫條圖
df2.plot.barh(stacked=True)
```

### 請用鐵達尼的資料檔計算不同票價的存活率？並用其視覺化呈現？

```python
# 票價：fare 存活：alive(yes, no)
df3 = pd.crosstab(titanic["fare"], titanic["alive"])
df3
```

```python
df3_1 = pd.crosstab(titanic["fare"], titanic["alive"], normalize=True)
df3_1
```

```python
df3_1.drop(columns=['no'], axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(60,6)) # 設定圖表大小為10x6英寸
ax = df3_1.plot.bar(stacked=True, ax=fig.add_subplot(111))
plt.xticks(rotation=90) # 旋轉x軸名稱方向
plt.subplots_adjust(bottom=0.2) # 調整圖表底部的空間
plt.show()
```

### 請綜合評估三個因素，鐵達尼事件中哪一類身份的人存活率最高？

```python
#性別 sex、票價 fare、艙等 pclass v.s 存活率 alive
df4 = pd.crosstab(index=[titanic["sex"], titanic["fare"], titanic["pclass"]], columns=titanic["alive"])
df4
```

```python
df4_1 = pd.crosstab(index=[titanic["sex"], titanic["fare"], titanic["pclass"]], columns=titanic["alive"], normalize=True)
df4_1
```

```python
df4_1.drop(columns=['no'], axis=1, inplace=True) # 刪除 no 列
```

```python
# 將存活率為零的項目刪除
import pandas as pd
import numpy as np

# 刪除值為零的行
df4_1 = df4_1.loc[(df4_1 != 0).any(axis=1)]
# 顯示交叉表格
print(df4_1)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(60,6)) # 設定圖表大小為10x6英寸
ax = df4_1.plot.bar(stacked=True, ax=fig.add_subplot(111))
plt.xticks(rotation=90) # 旋轉x軸名稱方向
plt.subplots_adjust(bottom=0.2) # 調整圖表底部的空間
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 只留下 yes 的值介於 75% 至最大值之間的項
q75, qmax = np.percentile(df4_1['yes'], [75, 100])
df4_1 = df4_1[df4_1['yes'].between(q75, qmax)]

fig = plt.figure(figsize=(60,6)) # 設定圖表大小為10x6英寸
ax = df4_1.plot.bar(stacked=True, ax=fig.add_subplot(111))
plt.xticks(rotation=90) # 旋轉x軸名稱方向
plt.subplots_adjust(bottom=0.2) # 調整圖表底部的空間
plt.show()
```
