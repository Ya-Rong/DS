- 請問WIND_SPEED 、TEMP、RAIN 、PH_RAIN、Humidity 、RAIN_COND以下變項預測PM25，並列出回歸方程式？
- 請用SO2、CO、O3、PM10 、PM25 、Nox、NO 、NO2、THC、NMHC、CH4繪製集群熱力圖，並將相關係數標示於圖中。

```python
import statsmodels.api as sm

# 提取所需變數
variables = ['WIND_SPEED', 'TEMP', 'RAIN', 'PH_RAIN', 'Humidity', 'RAIN_COND']

subset_data = df[variables]
subset_data = sm.add_constant(subset_data) # 添加常數項
# 建立線性迴歸模型
model = sm.OLS(df['PM25'], subset_data)

# 建立模型擬合
results = model.fit()

# 顯示迴歸方程式
print('迴歸方程式：')
print(results.summary())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 讀取數據文件
df = pd.read_excel('2017_PM25.xlsx')

# 提取所需變數
variables = ['SO2', 'CO', 'O3', 'PM10', 'PM25', 'Nox', 'NO', 'NO2', 'THC', 'NMHC', 'CH4']

subset_data = df[variables]

# 計算變數間的關聯性
correlation_matrix = subset_data.corr()

# 繪製 clustermap
sns.clustermap(correlation_matrix, cmap='coolwarm', annot=True, fmt='.2f')

# 設定圖形標題
plt.title('Clustermap of Variabkes')

# 顯示圖形
plt.show()
```
