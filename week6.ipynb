# Pandas資料轉換、資料統計、匯出儲存檔案

```python
import pandas as pd
df1 = pd.read_csv('hellopython2.csv')
```

## Pandas資料轉換

```python
print(df1)
print("===")
print(df1.info())
```

```python
# 數值轉換：字串轉數值
df1['gender'] = df1['gender'].map({'male':1, 'female':0})
# 變數['行標籤'] = 變數['行標籤'].map({'字串1':新值1, '字串2':新值2})
print(df1)
print('===')
print(df1.info())
```

```python
# 數值轉換：數值轉數值
df1['gender'] = df1['gender'].map({1:1, 0:2})
# 變數['行標籤'] = 變數['行標籤'].map({原值1:新值1, 原值2:新值2})
print(df1.head())
print('===')
print(df1.info())
```

```python
# 資料類型屬性、字串轉換
df2 = pd.read_csv('hellopython2.csv')
print(df2)
print(df2['gender'].str.title().head())
# 字串的大小寫變更，若要讓字串第一個字為大寫，df['STR'].str.title()
print(df2['gender'].str.lower().head())
# 字串全部變成小寫，df['STR'].str.lower()
print(df2['gender'].str.upper().head())
# 讓字串全部變成大寫，df['STR'].str.upper()
df2['prefer'] = df2['prefer'].str.replace('puma','彪馬')
# 字串變更，df['STR']=df['STR'].str.replace('原有字串', '欲改變成的字串')
print(df2['prefer'].head())
```

```python
# 資料篩選
df3 = pd.read_csv('hellopython2.csv')
print(df3)
print('===')
print(df3[df3['wt'] > 70])
# 利用中括號[]來指定存取欄位並設定條件進行資料篩選：變數[變數['欄標籤名稱']]
```

```python
# 在行標籤名稱中，找出包含特定值的資料集
# isin() 
print(df3)
print('===')
print(df3[df3['prefer'].isin(['adidas'])])
# 變數[變數['行標籤'].isin([選取特定值])]
```

```python
# 資料排序
# 列索引排序
# .sort_index()
df4 = pd.read_csv('hellopython2.csv')
print('===')
new_df = df4.sort_index(ascending=False)
print('遞減排序：')
print(new_df.head())
# 遞增排序：變數.sort_index(ascending = True)
# 遞減排序：變數.sort_index(ascending = False)
```

```python
# 資料排序
# 標籤值排序
print(df4.head())
print('===')
new_df2 = df4.sort_values(['wt'], ascending = True)
print('遞增排序')
print(new_df2.head())
print('===')
new_df3 = df4.sort_values(['wt'], ascending = False)
print('遞減排序')
print(new_df3.head())
```

### 資料合併

```python
# 資料合併
df5 = pd.read_csv('hellopython2.csv')
df6 = pd.read_csv('hellopython3.csv')
print(df5)
print('===')
print(df6)
```

```python
# 從資料尾部合併
# 行標籤相同，增加資料筆數
# 變數1.append(變數2)
print('輸出合併資料：')
append_df1 = df5.append(df6)
print('===')
print('重置列索引並輸出倒數五筆資料：')
riappend_df1 = append_df1.reset_index(inplace=False, drop=True)
# reset_index()：重置列索引，以利後續資料處理。
print(riappend_df1.tail())
```

```python
# 從資料標籤合併
# merge() ：增加行標籤的方式合併。
mergeleft = pd.read_csv('mergeleft.csv')
mergeright = pd.read_csv('mergeright.csv')                        
print('原始變數1:\n', mergeleft)
print('===')
print('原始變數2:\n', mergeright)
```

```python
# 指派變數 = pd.merge(原始變數1, 原始變數2, on = 'NO', how = 'outer')
merge_df1 = pd.merge(mergeleft, mergeright, on = 'NO', how = 'outer')
print(merge_df1)
```

## Pandas 資料統計

```python
import pandas as pd
```

```python
# 描述統計
df = pd.read_csv('py01nsc2011.csv')
print(df.columns)
```

```python
# 特定描述統計函式
print('最大值', df['Sportyear'].max())
print('最小值', df['Sportyear'].min())
print('平均數', df['Sportyear'].mean())
print('標準差', df['Sportyear'].std())
print('次數', df.value_counts(['Sportyear']))
```

```python
# 使用描述統計函式
# describe()
print(df['Gender'].describe())
print('===')
print(df['Sportyear'].describe())
```

```python
# 全部行標籤描述統計
# (describe())：描述統計
print(df[:].describe())
# 選擇全部行標籤資料函式 (df[:]) 
```

```python
# 新增行標籤資料
x1 = ['SS01','SS02','SS03','SS04']
df['SST1'] = df[x1].sum(axis = 'columns')
# 新增一行標籤的函式為 (axis = 'columns') 或寫為 (axis = 1)
# 加總行標籤函式為：變數[加總行標籤變數].sum(axis = 'columns')
# 新增行標籤函式為：變數['新增行標籤']
print(df['SST1'].describe())
```

```python
print(df[x1].mean())
df['SST1M'] = df[x1].mean(axis = 1)
print('輸出SST1平均：', df['SST1M'].mean())
print('===')
print(df['SST1M'].describe())
```

## Pandas匯出儲存檔案

```python
# 資料儲存函式：df.to_csv('檔名.csv') 
df.to_csv('py01nsc2011_1.csv') 
df1 = pd.read_csv('py01nsc2011_1.csv')
print(df1.columns)
```

## 數值資料分析與視覺化：Numpy及matplotlib

##### 圖示類型與功能

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('python2011nsc.csv')
```

```python
df
```

### 折線圖 (line chart)

```python
# 指派變數.plt.plot(參數)
plt.plot(df['Grade'])
plt.show()
```

### 長條圖 (bar/barh chart)

```python
x = [1, 0]
gender = [389, 141]
plt.bar(x, gender, color = 'b')
plt.xticks(x, ['male', 'female'])
plt.show()
```

```python
# pandas plot in bar chart

df['Gender'].value_counts().plot.bar(rot=0, colormap='summer')
# 計次函式value_counts()
# 繪製長條圖函式「變數.plot.bar()
```

### 圓餅圖 (pie chart)

```python
# 設定類別軸標籤，autopct 設到小數第一位，'equal'讓圓餅圖比例一致
size = df['Gender'].value_counts() # value_counts()計算個數並指派為變數size
labels = ['male', 'female'] # 設定圓餅圖類別標籤labels
plt.pie(size, labels = labels, autopct = '%.2f%%') # 圓餅圖函式plt.pie() 
# 2f後面出現兩個 %% 代表圓餅圖以百分比方式呈現
# 只有一個 % 就不會是百分比方式
plt.axis('equal')
plt.show()
```

### 直方圖 (histogram)

```python
# 「matplotlib」模組：
plt.hist(df['Grade'], 10) # plt.hist(變數, 分組數量)
plt.show()
# 「Pandas」模組繪製直方圖函式則是「變數.plot.hist()」
```

### 散布圖 (scatter chart)

```python
plt.scatter(df['ss'], df['achievement'], c='darkblue', alpha=.4)
# plt.scatter(變數1, 變數2, color='顏色', alpha=.5)
plt.show()
# Pandas散布圖： 變數.plot.scatter(x= '變數1', y= '變數2')
```

## 圖示調整

```python
# 圖示標籤與圖示儲存
plt.xlabel('social support') # X軸標籤名稱：plt.xlabel('x軸標籤名稱')
plt.ylabel('achievement') # Y軸標籤名稱：plt.ylabel('y軸標籤名稱')
plt.title('Scatter of ss and achieve') # 圖示名稱：plt.title('圖示名稱')
plt.scatter(df['ss'], df['achievement'], c='darkblue', alpha=.4)
plt.savefig('scatter of ss and achieve') # 儲存匯出圖示：plt.savefig('儲存圖示名稱')
plt.show()
```

[標號 (marker)官方款式](https://matplotlib.org/stable/api/markers_api.html)

```python
# 標號 (marker)
plt.plot([0,1,4],[1,4,7], marker='o') # 'o'：circle
plt.show()
```

```python
# 線的樣式 (linestyle)
fig, ax = plt.subplots(1,3) # fig是整個圖表窗口，而ax是圖表中的子圖。通過控制fig和ax的屬性，可以實現對整個圖表和子圖的控制，從而實現各種不同的繪圖效果
# plt.subplots(1,3) 1：row, 3：column
ax[0].plot([1,3,5,7],[2,4,3,5])
ax[1].plot([1,3,5,7],[2,4,3,5], linestyle='--')
ax[2].plot([1,3,5,7],[2,4,3,5], linestyle=':')
```

```python
# 線寬 (linewidth)
# 線的寬度調整，只能在折線圖中使用
fig, ax = plt.subplots(1,2)
ax[0].plot([1,2,3,4],[2,4,2,4])
ax[1].plot([1,2,3,4],[2,4,2,4], linewidth=10)
```

```python
# 標籤 (label)
# 呈現圖例，製作標籤與命名，再使用函式legend()
fig, ax = plt.subplots(1)
ax.plot([1,3,5,7],[2,5,2,4], label = 'line1')
ax.plot([1,2,4,8],[3,7,5,2], label = 'line2', linestyle='--')
ax.legend()
```

```python
# 透明度 (alpha)
fig, ax = plt.subplots(1,2)
ax[0].plot([1,3,5,7],[1,3,2,4])
ax[1].plot([1,3,5,7],[5,6,4,1], alpha=0.5)
```
